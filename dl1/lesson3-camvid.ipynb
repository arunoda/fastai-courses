## Let's create the DataBunch

Before we do that, let's try to get a new size of the image which is smaller but quite okay to work with. <br/>
Later on we can use the full size to get more accuracy.

```python
img = open_image(images[0])
```

```python
src_size = np.array(img.size); src_size
```

```python
new_size = src_size//2; new_size
```

```python
img.resize((3, 240, 320))
```

```python
# load data
d = SegmentationItemList.from_folder(path/'images')
# split them
d = d.split_by_fname_file(path/'valid.txt')
# create labels
d = d.label_from_func(get_mask_path, classes=codes)

# get the databunch (We may need to change this bs flag is we ran out of GPU memory)
db = (d
      .transform(get_transforms(), size=new_size, tfm_y=True)
      .databunch(bs=16)
      .normalize(imagenet_stats))
```

```python
db.show_batch(2, figsize=(6,6), dstype=DatasetType.Valid)
```

Since we are using the `SegmentationItemList`, databunch is render both image and the mask on top of that as shown above.

(Try removing `tfm_y=True` from the `transform()` and see how it behaves)

## Initial Training

```python
# In the paper related to this dataset, they asked to remove Void pixels 
# when calculating the accuracy.
codeMap = {v:k for k,v in enumerate(codes)}
codeMap['Void']
```

```python
def acc_camvid(input, target):
    target = target.squeeze(1)
    mask = target != codeMap['Void']
    return (input.argmax(dim=1)[mask]==target[mask]).float().mean()
```

```python
learn = Learner.create_unet(db, models.resnet34, metrics=acc_camvid)
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(10, slice(1e-2))
```

```python
learn.save('res34-freeze')
```

### Unfreeze and Proceed

```python
learn.load('res34-freeze');
```

```python
learn.unfreeze()
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(10, slice(1e-6, 1e-4))
```

```python
learn.save('res34-unfreeze')
```
