## Loading Data

All the datsets related to this notebook can be found here: https://goo.gl/cF6uWE

Currently it's not possible to load my dataset using `untar_data` due to a [bug](https://github.com/fastai/fastai/issues/983) in fastai.<br/>
So, I've to come up with the following utility function to download my dataset.

```python
dataset_url = "https://github.com/arunoda/fastai-courses/releases/download/fastai-vehicles-dataset/fastai-vehicles.tgz"
```

```python
import os
import subprocess

def exec(commands, cwd=None, show_output=False):
  FNULL = open(os.devnull, 'w')
  if show_output == True:
    cp = subprocess.run(commands, cwd=cwd, universal_newlines=True)
  else:
    cp = subprocess.run(commands, cwd=cwd, stdout=subprocess.PIPE, stderr=subprocess.STDOUT, universal_newlines=True)
  
  if cp.returncode != 0:
    raise Exception(cp.stdout)

def get_data(name, url):
  homedir = os.getenv('HOME')
  datadir = f'{homedir}/userdata'
  tmpdir = f'/tmp/{name}'
  tarfile = f'{tmpdir}/data.tgz'
  finaldir = f'{datadir}/{name}'

  if os.path.exists(finaldir):
    return Path(finaldir)

  exec(['rm', '-rf', tmpdir])
  exec(['mkdir', '-p', tmpdir])
  exec(['curl', '-L', url, '--output', tarfile], show_output=True)
  exec(['tar', 'xzf', tarfile], cwd=tmpdir)
  exec(['rm', tarfile])

  exec(['rm', '-rf', finaldir])
  exec(['mkdir', '-p', datadir])
  exec(['mv', tmpdir, datadir])
  
  return Path(finaldir)
```

```python
path = get_data("fastai-vehicles", dataset_url); path
```

```python
path.ls()
```

```python
path_img = path/'images'
fnames = get_image_files(path_img)
len(fnames)
```

```python
fnames[:5]
```

Now we've some idea on the filenames of this dataset.<br/>
Let's try to create a DataBunch from these images.

```python
pattern = r' *([^/]+)-\d+.jpg$'
data = ImageDataBunch.from_name_re(path_img, fnames, pattern, ds_tfms=get_transforms(), size=224)
data.normalize(imagenet_stats)
```

```python
print(len(data.classes))
data.classes
```

Let's see some random images from this dataset.

```python
data.show_batch(3, (10, 10))
```

```python
# This is a helper function which will help us later on
def find_lr(l):
    l.lr_find();
    l.recorder.plot();
```

## Case1: use resnet34 for training

Let's try to train a classifier with resnet34

```python
learn1 = create_cnn(data, models.resnet34, metrics=error_rate)
```

```python
learn1.fit_one_cycle(3)
```

```python
learn1.save('vehicles-l1-freeze')
```

```python
find_lr(learn1)
```

```python
learn1.unfreeze()
```

```python
learn1.fit_one_cycle(2, slice(1e-5, 1e-2))
```

**Here, unfreeze results are pretty good.** <br/>

## Inspect Result

Let's see what has happened to our model.

```python
inspect = ClassificationInterpretation.from_learner(learn1)
```

```python
inspect.plot_top_losses(9)
```

```python
inspect.plot_confusion_matrix()
```

```python
inspect.most_confused(1)
```

## Case2: use resnet50

```python
learn2 = create_cnn(data, models.resnet50, metrics=error_rate)
learn2.fit_one_cycle(3)
```

```python
find_lr(learn2)
```

```python
learn2.unfreeze()
learn2.fit_one_cycle(2, slice(1e-5, 1e-3))
```

**Notes**

The improvement we got here is similar to what we've got with resnet34. <br/>
So, changing the model doesn't improve the result significantly.

##  Case3: Use square images 

Usually an image of a vehicle is a rectangle. fastai needs a square image, so it does a center crop. <br/>
I think that's a way we lose some details of those images.<br/>
To get rid of that, I'm using a different dataset which is already resized as squre images.<br/>
(By adding some white spaces to top and bottom of the image)

```python
dataset_url_sq = 'https://github.com/arunoda/fastai-courses/releases/download/fastai-vehicles-dataset/fastai-vehicles-square.tgz'
path_sq = get_data('fastai-vehicles-square', dataset_url_sq) 
```

```python
path_img_sq = path_sq/'images'
fnames_sq = get_image_files(path_img_sq)
```

```python
pattern = r' *([^/]+)-\d+.jpg$'
data_sq = ImageDataBunch.from_name_re(path_img_sq, fnames_sq, pattern, ds_tfms=get_transforms(), size=224)
data_sq.normalize(imagenet_stats)
```

```python
data_sq.show_batch(3, (6, 6))
```

```python
learn3 = create_cnn(data_sq, models.resnet34, metrics=error_rate)
learn3.fit_one_cycle(3)
```

```python
find_lr(learn3)
```

```python
learn3.unfreeze()
learn3.fit_one_cycle(2, slice(1e-6, 1e-3))
```

**Notes**<br/>

Even with the square images, the result is worst than before.<br/>
So, our original **resnet34** version is still better.

## Case4: Multiple Crops

Let's try to use the original dataset with the resnet34 but with 3 different crops from each images.<br/>
Here's an example:

![](https://user-images.githubusercontent.com/50838/47656230-b3169d00-dbb4-11e8-8468-1e09cea2a32c.png)

Hopefully, this will give us some better result.

```python
path_crops = get_data('fastai-vehicles-crops', 'https://github.com/arunoda/fastai-courses/releases/download/fastai-vehicles-dataset/fastai-vehicles-crops.tgz')
```

```python
img_path_crops = path_crops/'images'
```

```python
fnames_crops = get_image_files(img_path_crops)
```

```python
fnames_crops[:5]
```

```python
pattern_crop = r'( *[^/]+)-[east|west|center]+-\d+.jpg$'
```

```python
data_crops = ImageDataBunch.from_name_re(img_path_crops, fnames_crops, pattern_crop, ds_tfms=get_transforms(), size=224)
```

```python
data_crops.normalize(imagenet_stats)
```

```python
data_crops.show_batch(3, (10, 10))
```

```python
learn4 = create_cnn(data_crops, models.resnet34, metrics=error_rate)
```

```python
learn4.fit_one_cycle(3)
```

```python
learn4.save('vehicles-l4-freeze')
```

```python
find_lr(learn4)
```

```python
learn4.unfreeze()
```

```python
learn4.fit_one_cycle(2, slice(1e-5, 1e-3))
```

**Notes**

This is a really good improvement. By providing more data on different parts of the image, we improve the accuracy of our classifier.
